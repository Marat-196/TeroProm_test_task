```python
import pandas as pd
```

```python
df_1 = pd.read_excel('data/Список товаров.xlsx')
```

```python
for i in range(len(df_1)):
    df_1['Наименование'].loc[i] = df_1['Наименование'].str.split('.').loc[i][0]
```

```python
df_1['tag'] = None
or i in range(len(df_1)):
    df_1['tag'].loc[i] = df_1['Наименование'].str.split()[i][0][:-1]
```
